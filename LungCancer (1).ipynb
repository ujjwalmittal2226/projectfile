```python
df_new = df.drop(columns = ['Gender','Weight Loss', 'Shortness of Breath', 'Wheezing',
       'Swallowing Difficulty', 'Clubbing of Finger Nails', 'Frequent Cold', 'Snoring'])
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df_new['Level'] = le.fit_transform(df_new['Level'])
```

```python
df_new.head()
# low -> 1
# medium -> 2
# high -> 0
```

```python

plt.figure(figsize=(15,15))
sns.heatmap(df_new.corr(),annot=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

X = df_new.drop('Level',axis=1)
y = df_new.Level

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=44)
```

```python
X_train.head()

```

```python
y_train.head()
```

## XGBOOST

```python
xgb = XGBClassifier( learning_rate=0.1)
xgb.fit(X_train,y_train)
```

```python
xgb.score(X_test,y_test)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import confusion_matrix

print(mean_squared_error(y_test,xgb.predict(X_test)))
print(confusion_matrix(y_test,xgb.predict(X_test)))
```

```python
plt.figure(figsize=(7,7))
sns.heatmap(confusion_matrix(y_test,xgb.predict(X_test)),annot=True)
```
